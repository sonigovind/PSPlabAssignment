```python
#discrete random variable of throwing a dice
import scipy.stats as stats
```

```python
die=stats.randint(1,7)
```

```python
r=range(1,7)
```

```python
prob=[die.pmf(i) for i in r]
```

```python
import numpy as np
sample=np.random.randint(1,7,size=50)
plt.hist(sample,bins=50,label='sample1',alpha=0.5)
```

```python
# for continous variable
from numpy.random import uniform
uniform(low=-3,high=3,size=50)
```

```python
s=uniform(low=-3,high=3,size=50)
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.hist(s,density=True, label='sample2',alpha=0.5)
```

```python
#Comparing histogram
plt.hist(sample,bins=50,alpha=0.5 ,label='sample1')
plt.hist(s,density=True,bins=50,alpha=0.5,label='sample2')

plt.legend(loc='best')
plt.show()
```
